```python
import seaborn as sns
import matplotlib.pyplot as plt
iris =sns.load_dataset("iris")
print(iris)
sns.scatterplot(x="sepal_length",y="species",data=iris,label="iris")
plt.title("sepal_length vs sepal_width ")
plt.xlabel("sepal_length($)")
plt.ylabel("sepal_width ($)")
plt.legend(loc='best')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris =sns.load_dataset("iris")
print(iris)
sns.barplot(x="sepal_length",y="species",data=iris,label="iris")
plt.title("sepal_length vs sepal_width ")
plt.xlabel("sepal_length($)")
plt.ylabel("sepal_width ($)")
plt.legend(loc='best')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris =sns.load_dataset("iris")
print(iris)
sns.boxplot(x="sepal_length",y="sepal_width",data=iris)
plt.title("sepal_length vs sepal_width ")
plt.xlabel("sepal_length($)")
plt.ylabel("sepal_width ($)")
plt.legend(loc='best')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris =sns.load_dataset("iris")
print(iris)
sns.violinplot(x="sepal_length",y="sepal_width",data=iris)
plt.title("sepal_length vs sepal_width ")
plt.xlabel("sepal_length($)")
plt.ylabel("sepal_width ($)")
plt.legend(loc='best')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as mpt
tips=sns.load_dataset("tips")
print(tips)
sns.jointplot(x='total_bill',y='tip',data=tips,label='vikas')
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tips ($)")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as mpt
tips=sns.load_dataset("tips")
print(tips)
sns.lmplot(x='total_bill',y='tip',data=tips)
g=sns.FacetGrid(tips,col="tip")
g.map(sns.histplot,"day")
plt.subplots_adjust(top=0.8)
g.fig.suptitle()
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tips ($)")
plt.legend()
plt.show()
```
